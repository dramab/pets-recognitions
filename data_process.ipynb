## 将图片复制到指定文件夹

```python
class_id = []
place = r"E:\py_code\大创\Pets-Face-Recognition-main\pets_datasets\oxford-iiit-pet\images"
path = os.listdir(place)
```

```python
for i in path:
    #删除.mat结尾的文件索引
    if '.mat' in i:
        path.remove(i)
len(path)
```

```python
path[0].split('.jpg')[0]
```

```python
path1 = list(map(lambda x : x.rstrip('_0123456789.jpg'),path))
path1
```

```python
path2 = list(set(path1))
len(path2)
```

```python
dic = {'a':'x'}
'x' in dic,'a' in dic
```

```python
class_id = {}
for idx in range(len(path)):
    
    file_name = path1[idx]
    if file_name not in class_id:
        class_id[file_name] = [path[idx]]
    else:
        class_id[file_name].append(path[idx])
len(class_id)
```

```python
a = [1,2]
b = [2]
for i in b:
    a.remove(i)
a
```

```python
import random
train_dic = {}
val_dic = {}
for cls in class_id:
    val_dic[cls] = random.sample(class_id[cls],5)
    
    train_dic[cls] = list(set(class_id[cls])-set(val_dic[cls]))
val_dic
```

```python
train_path = os.path.join(dataset,'train')
val_path = os.path.join(dataset,'val')
os.makedirs(train_path,exist_ok = True)
os.makedirs(val_path,exist_ok = True)
for val_pic in val_dic:
    #生成测试集
    os.makedirs(os.path.join(dataset,'val',val_pic))
    for pic in val_dic[val_pic]:
        shutil.copy(os.path.join(place,pic),os.path.join(dataset,'val',val_pic))

for train_pic in train_dic:
    #生成测试集
    os.makedirs(os.path.join(dataset,'train',train_pic))
    for pic in train_dic[train_pic]:
        shutil.copy(os.path.join(place,pic),os.path.join(dataset,'train',train_pic))
```
